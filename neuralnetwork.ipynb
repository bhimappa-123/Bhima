```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(3,9,1000)
```

```python
x
```

```python
np.random.seed(3)
y = np.sin(x) + np.random.normal(0,0.3,1000)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
model = MLPRegressor((100,20))
```

```python
xx = x.reshape(-1,1)
```

```python
model.fit(xx,y)
```

```python
yp = model.predict(xx)
```

```python
plt.scatter(x,y)
plt.plot(x,yp,"*r")


```
